```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("data/processed/forecast.csv")
dc1_racks = df[(df["DataCentre"]=="DC1") & (df["Metric"]=="Total_Contracted")]

plt.plot(dc1_racks["ds"], dc1_racks["yhat"], label="Forecast")
plt.fill_between(dc1_racks["ds"], dc1_racks["yhat_lower"], dc1_racks["yhat_upper"], alpha=0.2)
plt.legend()
plt.show()
```
